```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iou = pd.read_csv('IoU.csv')
```

```python
iou
```

```python
l1=np.asarray(iou['val1'].tolist())
l2=np.asarray(iou['val2'].tolist())
l4=np.asarray(iou['val4'].tolist())
l6=np.asarray(iou['val6'].tolist())
l8=np.asarray(iou['val8'].tolist())
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)

ax.plot(np.arange(100), l1, label='1 layer')
ax.plot(np.arange(100), l2, label='2 layer')
ax.plot(np.arange(100), l4, label='4 layer')
ax.plot(np.arange(100), l6, label='6 layer')
ax.plot(np.arange(100), l8, label='8 layer')

ax.legend()
ax.set_xlabel('Epoch number')
ax.set_ylabel('mean IoU')
fig.tight_layout()
plt.show()
fig.savefig('layer.pdf')
```

```python
relu = np.asarray(iou['val2'].tolist())
elu = np.asarray(iou['valelu'].tolist())
selu = np.asarray(iou['valselu'].tolist())
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)

ax.plot(np.arange(100), relu, label='ReLU')
ax.plot(np.arange(100), elu, label='ELU')
ax.plot(np.arange(100), selu, label='SeLU')

ax.legend()
ax.set_xlabel('Epoch number')
ax.set_ylabel('mean IoU')
fig.tight_layout()
plt.show()
fig.savefig('activation_function.pdf')
```

```python
adam=np.asarray(iou['val2'].tolist())
rms=np.asarray(iou['valrms'].tolist())
sgd=np.asarray(iou['valsgd'].tolist())
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)

ax.plot(np.arange(100), adam, label='Adam')
ax.plot(np.arange(100), rms, label='RMSProp')
ax.plot(np.arange(100), sgd, label='SGD')

ax.legend()
ax.set_xlabel('Epoch number')
ax.set_ylabel('mean IoU')
fig.tight_layout()
plt.show()
fig.savefig('Optimizer.pdf')
```

```python
dp1 = np.asarray(iou['valdp1'].tolist())
dp3 = np.asarray(iou['valdp3'].tolist())
dp5 = np.asarray(iou['valdp5'].tolist())
dp7 = np.asarray(iou['valdp7'].tolist())
dp0 = np.asarray(iou['val2'].tolist())
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)

ax.plot(np.arange(100), dp1, label='0.1')
ax.plot(np.arange(100), dp3, label='0.3')
ax.plot(np.arange(100), dp5, label='0.5')
ax.plot(np.arange(100), dp7, label='0.7')
ax.plot(np.arange(100), dp0, label='No dropout')

ax.legend()
ax.set_xlabel('Epoch number')
ax.set_ylabel('mean IoU')
fig.tight_layout()
plt.show()
fig.savefig('dropout.pdf')
```

```python
l1tr = np.asarray(iou['tr1'].tolist())
l1val = np.asarray(iou['val1'].tolist())
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)

ax.plot(np.arange(100), l1tr, label='train')
ax.plot(np.arange(100), l1val, label='valid')
ax.legend()
ax.set_xlabel('Epoch number')
ax.set_ylabel('mean IoU')
fig.tight_layout()
plt.show()
fig.savefig('layer1.pdf')
```
